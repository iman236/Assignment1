```python
import pandas as panda
from matplotlib import pyplot as plt
```

```python
dataframe=panda.read_csv("Ruter_data.csv", delimiter=";")
```

```python
dataframe.head(50)
```

```python
dataframe.shape
```

```python
plt.hist(dataframe.Kjøretøy_Kapasitet, bins=5, rwidth=0.6)
plt.xlabel('Kapasitet')
plt.ylabel('Count')
plt.show()
```

```python
dato=dataframe[dataframe.Dato=="10/08/2020"]
dato.head(100)
```

```python
plt.hist(dato.Område, bins=4, rwidth=0.8)
plt.title("Activity in each area 11/08/2020")
plt.xlabel("Område")
plt.ylabel("Antall")
plt.show()
```

```python
dato1=dataframe[dataframe.Dato=="11/08/2020"]
dato1.head(100)
```

```python
plt.bar(dato.Område, , data=dato, color="blue")
plt.bar(dato1.Område, , data=dato1, color="green")
plt.title("Activity in each area 10-11/08/2020")
plt.xlabel("Område")
plt.ylabel("Antall")
plt.show() 
#comparing activity in each 'område' for two consecutive days
```
